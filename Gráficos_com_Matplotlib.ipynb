#Comando para visualização básico:
 
##I: Para fazer alguma visualização, primeiro é necessário executar o comando plt.figure();
##II: Após inicializar nossa figura, ainda vazia, para colocar um eixo do tamanho da figura (máximo), devemos executar fig.add_axes([0, 0, 1, 1]);
##III: No comando eixo.plot(), deve-se sempre passar dois argumentos. Um deles é um DataFrame, contendo os dados a serem visualizados, e outro é o título de nossa visualização;
##IV: Título, nomes dos eixos, e outros detalhes podem ser acrescentados depois de executarmos o comando que faz a visualização (.plot())

## Mais de um eixo na mesma visualização

```
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])
eixo2 = fig.add_axes([0.1,0.5,0.4,0.4])
eixo.grid(True)
eixo.plot(dados['data'],dados['temperatura'], color= 'g' )
eixo.set_title("Temperatura no momento",fontsize= 25, pad=7)
eixo.legend(['temperatura'],loc = 'best',fontsize= 15)
eixo2.plot(dados['data'],dados['temperatura'],color = 'b')
eixo2.set_title('Temperatura no momento',fontsize= 15,pad = 10)
```

# Marcando o ponto máximo e mínimo do gráfico

```
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(dados['data'],dados['temperatura'],color='g')
eixo.set_title('Temperatura no momento',pad = 20, fontsize = 20)
eixo.set_xlabel('Data',fontsize = 20)
eixo.set_ylabel('Temperatura',fontsize = 20)
eixo.legend(['temperatura'],loc = 'lower right',fontsize = 15)
eixo.grid(True)

x1 = dados['data'][dados['temperatura'].idxmax()]
y1 = max(dados['temperatura'])

x2 = dados['data'][dados['temperatura'].idxmax()-7000]
y2 = max(dados['temperatura'])-5

eixo.annotate('Máximo',xy=(x1,y1),fontsize= 20,xytext = (x2,y2), arrowprops = dict(facecolor = 'k'))

x1 = dados['data'][dados['temperatura'].idxmin()]
y1 = min(dados['temperatura'])

x2 = dados['data'][dados['temperatura'].idxmin()-7000]
y2 = min(dados['temperatura'])+5

eixo.annotate('Mínimo',xy=(x1,y1),fontsize= 20,xytext = (x2,y2), arrowprops = dict(facecolor = 'k'))


eixo.axhline(max(dados['temperatura']),color = 'k',linestyle = '--')
eixo.axhline(min(dados['temperatura']),color = 'k',linestyle = '--')
```

# Visualizar os dias da semana e suas médias de temperaturas

```
temperaturas_por_dia_na_semana = dados.groupby('dia_da_semana')['temperatura'].mean()
temperaturas_por_dia_na_semana
```

```
nome_dias = ['Domingo','Segunda','Terça','Quarta','Quinta','Sexta','Sábado']
temperaturas_por_dia_na_semana = temperaturas_por_dia_na_semana[nome_dias]
temperaturas_por_dia_na_semana
```

```
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])
indice = range(len(temperaturas_por_dia_na_semana))
cores = ['black','blue','yellow','magenta','red','green','orange']
eixo.bar(indice, temperaturas_por_dia_na_semana, color = cores)
eixo.set_title('Temperatura por dia na semana',fontsize = 20)
eixo.set_xlabel('Dias da semana',fontsize = 20)
eixo.set_ylabel('Temperatura',fontsize= 20)
eixo.set_xticks(indice)
eixo.set_xticklabels(nome_dias)
```

# Pizza

```
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])
eixo.pie(temperaturas_por_dia_na_semana, labels=temperaturas_por_dia_na_semana.index)
```

## Ênfase no final de semana


```
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])
explodir = [0.1,0,0,0,0,0.1,0.1]
eixo.pie(temperaturas_por_dia_na_semana, labels=temperaturas_por_dia_na_semana.index,explode=explodir)
```

# Capítulo 4

```
iris = pd.read_csv('iris.csv')
```

```
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])


cores = {'Iris-setosa':'r','Iris-versicolor':'b','Iris-virginica': 'g'}
for especies in iris['espécie'].unique():
  temporario = iris[iris['espécie'] == especies]
  eixo.scatter(temporario['comprimento_sépala'],temporario['largura_sépala'], color = cores[especies])
eixo.set_title("gráfico de dispersão",fontsize = 20)
eixo.set_xlabel("Comprimento da sépala",fontsize = 20)
eixo.set_ylabel("Largura da sépala",fontsize = 20)
eixo.tick_params(labelsize = 15)
eixo.legend(cores,fontsize= 15)
```

# Boxplot com Matplotlib

```
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])
eixo.boxplot(iris['largura_pétala'])
eixo.set_title("Boxplot",fontsize= 15,pad = 10)
eixo.set_xticklabels(['largura_pétala'])
```

## fazendo boxplot agora com toda as variáveis, menos as das espécies

```
fig = plt.figure(figsize=(8,5))
eixo = fig.add_axes([0,0,1,1])

eixo.boxplot(iris.drop('espécie',axis = 1).values,patch_artist=True)
eixo.set_title("Boxplot",fontsize= 15,pad = 10)
eixo.set_xticklabels(iris.drop('espécie',axis =1).columns)
```

# Histogramas em Matplotlib

```
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])
eixo.hist(iris['comprimento_pétala'],bins = 20)
eixo.set_title("Histograma",fontsize= 20,pad = 10)
eixo.set_xlabel("Comprimento da pétala",fontsize = 15)
eixo.grid(True)
```

```
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])
mu, sigma = iris['comprimento_pétala'].mean(), iris['comprimento_pétala'].std()
eixo.hist(iris['comprimento_pétala'],bins = 20)
eixo.set_title("Histograma",fontsize= 20,pad = 10)
eixo.set_xlabel("Comprimento da pétala",fontsize = 15)
eixo.grid(True)
eixo.annotate('$mu = {0:.2f}$\n$sigma = {1:.2f}$'.format(mu,sigma), xy=(4.5,20),fontsize = 20)
```
